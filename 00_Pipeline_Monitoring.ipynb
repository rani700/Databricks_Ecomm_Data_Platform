```python
#Latest Runs (The "Pulse" of the system)
%sql
SELECT * FROM ecomm_data_project.audit.pipeline_logs 
ORDER BY timestamp DESC 
LIMIT 20;
```

```python
#Error Summary (The "Doctor's Report")
%sql
SELECT job_name, error_message, timestamp 
FROM ecomm_data_project.audit.pipeline_logs 
WHERE status = 'FAILED' 
ORDER BY timestamp DESC;
```

```python
#Data Volume Stats (The "Growth Tracker")
%sql
SELECT layer, SUM(records_processed) as total_rows 
FROM ecomm_data_project.audit.pipeline_logs 
GROUP BY layer;
```
